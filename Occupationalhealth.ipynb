```python
# Read the file

## Save the path
p ='/content/OccupationalHealth. (1).xlsx'
## Read and create the dataframe
d = pd.read_excel(p)
d.info()
```

# **3. Cleaning**

# Column names

```python
# Rename
d.rename(columns = {'DATA YEAR':'year'}, inplace = True)
d.rename(columns = {'ECONOMIC SECTOR':'economy'}, inplace = True)
d.rename(columns = {'AFFILIATED COMPANIES':'aff_comp'}, inplace = True)
d.rename(columns = {'Dependent affiliated workers':'workers_dep'}, inplace = True)
d.rename(columns = {'Independent affiliated workers':'workers_ind'}, inplace = True)
d.rename(columns = {'Diseases classified as occupational':'diseases'}, inplace = True)
d.rename(columns = {'Qualified work accidents':'accidents'}, inplace = True)

d.head()
```

## Replace columns

```python
d = d.replace('N.A.','Sin dato')
```

# **4. Indexing**

```python
# Duplicate the dataframe
d2 = d
```

```python
# Duplicate year as string / How to create a column
d2['year_str'] = d2['year'].astype(str)
d2.info()
```

```python
# Indexing
## Create the index

idx= pd.MultiIndex.from_product([d2['economy'].unique(),range(d2.year.min(), d2.year.max()+1)],
                                names = ['economy', 'year'])

## Attach the index
d2 = d2.set_index(['economy','year','aff_comp'])

d2.head()
```

## **5. Delete NAN**

```python
# Display if there are NaN
d2.isna()
```

```python
# Delete NaN
d2 = d2.dropna()
d2.info()
```

## **6. Descriptive statistics**

```python
# Basic
d2[['workers_ind','diseases','accidents']].describe().applymap(lambda x: "{:.1f}".format(x))
```

**Interpretation**

The information provided by the code shows us that :

Within sectors of the economy, there is an average of 19,469 people who are independently affiliated with an insurance company. Likewise, there is an average of 526 deaths and 9,455 accidents reported as occupational by these workers.
The economy with the most accidents reported 46,859 of these; The economy with the most deaths reported 10,639 of these.
The standard deviation of the data is high, which means that there is great variability between workers in each economy, their accidents and deaths respectively.



```python
# By index
d2[['workers_ind']].groupby('economy').describe().applymap(lambda x: "{:.1f}".format(x))
```

```python
tmp = d2.filter(like ='Domestic service', axis=0)
tmp = tmp.reset_index()
tmp.head()

pivot_pop = tmp.pivot(index = "year",
                    columns = "economy",
                    values = "workers_ind")
pivot_pop.head()
```

```python
domestic = d2.filter(like = 'Domestic service', axis=0)
domestic.head()
```

```python
fig2 = px.bar(domestic, x='year_str', y='workers_ind')

# Without the Legend

fig2.update_layout(showlegend = False)

# Center the title
fig2.update_layout(title_text = 'Domestic service workers',
                  title_font_size = 25,
                  title_x = 0.5)

# Modify the text in the axis
fig2.update_xaxes(title = 'Years', title_font_size = 20,
                 dtick= 'M1',
                 tickformat = '%b\n%Y')
fig2.update_yaxes(title = 'Workers', title_font_size = 20)
fig2.show()
```

# **Interpretation**

For this case, the domestic service economy was chosen, it is intended to represent that the number of independently affiliated workers in this sector had its highest record in 2019, having a next and important drop in 2020 of almost 95%. . But curiously, for the year 2021, the recovery was prompt, so the causal factors such as greater employability in this sector would have to be evaluated.



#**9. Another Graphs**

## Double axis

```python
## Libraries
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
## Create una figuera whit secondary axis

fig3 = make_subplots(specs=[[{"secondary_y": True}]])

## add each graph
fig3.add_trace(
    go.Scatter(x=domestic.year_str, y=domestic.workers_ind, name='workers_ind'),
    secondary_y=False,
)
fig3.add_trace(
    go.Scatter(x=domestic.year_str, y=domestic['diseases'], name='diseases'),
    secondary_y=True,
)

# Without the Legend

fig3.update_layout(showlegend = False)

# Center the title
fig3.update_layout(title_text = 'Deaths in domestic service',
                  title_font_size = 25,
                  title_x = 0.5)

# Modify the text in the axis
fig3.update_xaxes(title = 'Years', title_font_size = 18,
                 dtick= 'M1',
                 tickformat = '%b\n%Y')
fig3.update_yaxes(title = 'Workers', title_font_size = 18)

fig3.update_yaxes(title='Diseases', title_font_size= 18, secondary_y=True)

fig3.show()

```

**Interpretation**

The domestic service sector has had many regulations in recent years, so I find it interesting to analyze the number of deaths that independently affiliated workers have.
Particularly for the year with the highest registration of workers, the number of deaths was the lowest with 11 of these, while the year with the lowest registration of workers recorded the highest level of occupational deaths with 35 of them. It is important to review the causes of deaths in a sector like this, given that it is one of the sectors with the largest number of affiliated companies.
